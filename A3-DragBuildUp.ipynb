```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c=-0.0866
d=0.8099   # Roskam table 3.5
W_TO=60000  

Swet=10**(c+d*np.log10(W_TO))

print('Wetted Area = {} ft^2'.format(np.round(Swet,2)))


Cf=0.0026 #Roskam
a=-2.699+0.6*(2.699-2.5229)

f=10**(a+np.log10(Swet))

Sref=5000  # guess for now

CD0=f/Sref

print(CD0)
```

```python
CL=np.linspace(-3,3,61)
eclean=0.825
eTOFlap=0.775
eLFlap=0.725
eLGear=0.825
AR=12.3

CDclean=CD0+(CL**2)/(np.pi*eclean*AR)
CDTOflap=CD0+0.015+(CL**2)/(np.pi*eTOFlap*AR)
CDLFlap=CD0+0.065+(CL**2)/(np.pi*eLFlap*AR)
CDLGear=CD0+0.02+(CL**2)/(np.pi*eLGear*AR)

plt.figure(figsize=(8,8))
plt.plot(CL,CDclean,color='blue',label='Clean')
plt.plot(CL,CDTOflap,color='orange',label='Take Off Flaps')
plt.plot(CL,CDLFlap,color='purple',label='Landing Flaps')
plt.plot(CL,CDLGear,color='green',label='Landing Gear')
plt.legend()
plt.xlabel('CL')
plt.ylabel('CD')
plt.title('Drag Polar')
```
